# Cleaning Price column

```python
# Clean and convert 'Price' column (e.g., 'PKR\n4.75 Crore'  47500000)

def convert_price(price_str):
    try:
        # Remove "PKR", line breaks, and extra spaces
        clean_str = price_str.replace('PKR', '').replace('\n', ' ').strip()

        # Split into number and unit
        parts = clean_str.split()
        if len(parts) < 2:
            return None

        value = float(parts[0])
        unit = parts[1].lower()

        # Convert unit to numeric value
        if 'crore' in unit:
            return value * 10000000  # 1 crore = 10 million
        elif 'lakh' in unit:
            return value * 100000    # 1 lakh = 100 thousand
        elif 'thousand' in unit:
            return value * 1000
        else:
            return value
    except:
        return None

# Apply the conversion function
df['Price'] = df['Price'].astype(str).apply(convert_price)

# Display some converted prices
print("Cleaned numeric prices:")
print(df['Price'].head())
```

# Converting Area to Sqft

```python
# Converts area strings like '10 Marla', '1 Kanal', '128 Sq. Yd.' to sqft.

def convert_area(area_str):
    try:
        # Split into number and unit
        parts = area_str.strip().split()
        if len(parts) < 2:
            return None

        value = float(parts[0])
        unit = " ".join(parts[1:]).lower()  # Capture full unit like "sq. yd."

        # Define conversion rules
        if 'marla' in unit:
            return value * 225
        elif 'kanal' in unit:
            return value * 4500
        elif 'sqft' in unit or 'sq. ft' in unit:
            return value
        elif 'sq. yd' in unit or 'sq yd' in unit:
            return value * 9
        else:
            return None
    except:
        return None

# Apply the function to the Area column
df['Area'] = df['Area'].astype(str).apply(convert_area)

# Check sample output
print(df['Area'].head())
```

# FuzzyWuzzy
Standardized entries in City , Type and Province columns using fuzzywuzzy

```python
import fuzzywuzzy # pip install fuzzywuzzy
from fuzzywuzzy import process

def auto_standardize_column(df, column, threshold=90):
    from fuzzywuzzy import process

    unique_values = df[column].dropna().unique().tolist()
    standardized = {}

    for value in unique_values:
        if value in standardized.values():
            continue
        matches = process.extract(value, unique_values, limit=10)
        for match, score in matches:
            if score >= threshold:
                standardized[match] = value

    df[column] = df[column].map(standardized).fillna(df[column])
    print(f" Standardized {len(standardized)} entries in '{column}' using fuzzy matching.")

auto_standardize_column(df, 'City')
auto_standardize_column(df, 'Type')
auto_standardize_column(df, 'Province')


# Simple cleanup: Strip whitespace and standardize capitalization
df['Area Name'] = df['Area Name'].astype(str).str.strip().str.title()

```

# Boxplot Visualization for outliers

Outliers are removed from area and price by manually to only remove unrealistic, wrong data Because removing through IQR might delete originall data (Luxury properties)

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Reset index after any prior filtering
df.reset_index(drop=True, inplace=True)

# 1. Visualize original data (with outliers)
cols_to_plot = ['Price', 'Area', 'Bedrooms', 'Bathrooms', 'Property Age']

plt.figure(figsize=(15, 8))
for i, col in enumerate(cols_to_plot, 1):
    plt.subplot(2, 3, i)
    sns.boxplot(x=df[col])
    plt.title(f'Original: {col}')
plt.tight_layout()
plt.show()

# 2. Detect extreme outliers in 'Area' and 'Price' (for reporting only)
for col in ['Area', 'Price']:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Print number of extreme outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"\nExtreme outliers detected in '{col}': {len(outliers)}")

# 3. Define minimum and maximum thresholds based on real-world logic
min_area_sqft = 272.25        # 1 marla
max_area_sqft = 50000         # Very large but still possible (e.g., big farmhouses)
min_price = 100000            # Rs. 1 lakh (below this is unrealistic)
max_price = 1_000_000_000     # Rs. 100 crore (upper limit)

# 4. Apply filters to remove only extreme/unrealistic outliers
df = df[
    (df['Area'] >= min_area_sqft) & (df['Area'] <= max_area_sqft) &
    (df['Price'] >= min_price) & (df['Price'] <= max_price)
]

print(f"\n Shape after removing unrealistic outliers: {df.shape}")




# Reset index after any prior filtering
df.reset_index(drop=True, inplace=True)

```

# Feature Engineering
created age category column and luxury area column that flags a property as "luxury" or not based on certain conditions

```python
# Feature Engineering i have already created province and area from location column
# and property age from buit in year column now created age category column

df['Age Category'] = pd.cut(df['Property Age'], bins=[-1, 5, 15, 30, 100],
                            labels=['New (0-5 yrs)', 'Moderate (6-15)', 'Old (16-30)', 'Very Old (30+)'])

#Luxury Indicator is a new feature that flags a property as "luxury" or not based on certain conditions
luxury_areas = [
    'Dha Defence', 'Dha City Karachi', 'Clifton', 'Bath Island', 'Zamzama',
    'Frere Town', 'Kda Scheme 1', 'Sea View Apartments', 'Navy Housing Scheme Karsaz',
    'F-6', 'F-7', 'F-8', 'F-10', 'F-11', 'E-7', 'E-11', 'G-11', 'Bahria Town',
    'Park Enclave', 'Naval Anchorage', 'Multi Residencia & Orchards', 'Emaar Canyon Views',
    'Constitution Avenue', 'Dha 11 Rahbar', 'Askari', 'Johar Town', 'Bahria Orchard',
    'Lake City Meadows', 'Valencia Housing Society', 'Eden', 'Canal Gardens',
    'New Lahore City', 'Buch Executive Villas', 'Wapda Town', 'Askari Colony 2',
    'D Ground', 'Eden Gardens', 'Canal Road', 'Bahria Town Rawalpindi',
    'Askari 10', 'Askari 11', 'Askari 14', 'Taj Residencia'
]

df['Is Luxury'] = (
    (df['Price'] >= 200000000) &
    (df['Bedrooms'] >= 5) &
    (df['Area'] >= 4500) &
    (df['Area Name'].str.contains('|'.join(luxury_areas), case=False, na=False))
).astype(int)
```
